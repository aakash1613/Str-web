```python
import pandas as pd
import pickle
```

```python
df=pd.read_csv("RLData2.csv")
```

```python
df.head()
```

```python
df['furnishingstatus'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
aakash= LabelEncoder()
```

```python
df['furnishingstatus']=aakash.fit_transform(df['furnishingstatus'])
```

```python
df['furnishingstatus'].unique()
```

```python
df.head()
```

```python
x=df.iloc[:,1:7]
```

```python
y=df.iloc[:,0:1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
akash=LinearRegression()
```

```python
akash.fit(x_train,y_train)
```

```python
y_pred=akash.predict(x_test)
```

```python
y_pred
```
